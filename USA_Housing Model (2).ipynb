```python
def attribute_info(df_features):
    print(f'Description of features: \n\n {df_features.describe()} \n')
    print(f'Value counts: \n\n {df_features.value_counts()} \n')
    print(f'Unique features: \n\n {df_features.unique()} \n')
    print(f'Missing values: \n\n {df_features.isnull().sum()} \n')
    print(f'Missing values in %: \n\n {df_features.isnull().mean()} \n')
```

Evaluation Metrics

```python
#for training data
```

```python
def training_evaluation(actual, predicted):
    
    mse = mean_squared_error(actual, predicted)
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mae = mean_absolute_error(actual, predicted)
    r2 = r2_score(actual, predicted)
    
    
    print(f'MSE: {mse}')
    print(f'RMSE: {rmse}')
    print(f'MAE: {mae}')
    print(f'R2_Score: {r2}')
```

```python
training_evaluation(Y_train, Y_pred_train)
```

```python
#for testing data
```

```python
def test_evaluation(actual, predicted):
    
    mse = mean_squared_error(actual, predicted)
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mae = mean_absolute_error(actual, predicted)
    r2 = r2_score(actual, predicted)
    
    
    print(f'MSE: {mse}')
    print(f'RMSE: {rmse}')
    print(f'MAE: {mae}')
    print(f'R2_Score: {r2}')
```

```python
test_evaluation(Y_test, Y_pred_test)
```
